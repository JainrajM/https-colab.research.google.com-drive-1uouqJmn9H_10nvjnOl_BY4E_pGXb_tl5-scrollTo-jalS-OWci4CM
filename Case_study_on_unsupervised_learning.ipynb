```python
#import libraries
import pandas as pd
import numpy as np
import random as rd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/Wine_clust .csv')
data.head()
```

```python
sns.scatterplot(x='Alcohol',y='Proline',data=data)
```

```python
X=data
K=3
Centroids = (X.sample(n=K))
plt.scatter(X["Alcohol"],X["Proline"],c='black')
plt.scatter(Centroids["Alcohol"],Centroids["Proline"],c='red')
plt.xlabel('Total_Phenols')
plt.ylabel('Proline (In ten)')
plt.show()
```

```python
from sklearn.cluster import KMeans
K = 3
kmeans = KMeans(n_clusters=K, random_state=0)
X["Wine_clust"] = kmeans.fit_predict(X[["Alcohol", "Proline"]])
```

```python
X.head(10)
```

```python

# Get cluster centroids
centroids = kmeans.cluster_centers_
# Plot the data points and centroids
sns.scatterplot(data=X, x="Alcohol", y="Proline", hue="Cluster", palette="Set1")
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='x', s=100)
plt.xlabel('Total_Phenols')
plt.ylabel('Alcohol (In Ten)')
plt.show()
```

```python
inertia = kmeans.inertia_
print("Inertia:", inertia)
```
